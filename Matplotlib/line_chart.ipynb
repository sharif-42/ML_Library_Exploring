```python
import numpy as np
import matplotlib.pyplot as plt
```

# Line Chart
Line chart is strictly for lines but it takes a set of x y cordinates and plots them and each pair is connected to the next by line.

```python
x = np.linspace(0,15,100)
"""
linspace is a way of creating data. it takes 3 arguments called starting point, ending point
and how many data point is needed.
"""
y = np.sin(x)

plt.plot(x,y)
plt.xlabel('Time')
plt.ylabel('Function')
plt.title('Sine Wave')
plt.show()
```
